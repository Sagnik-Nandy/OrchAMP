```python
import pandas as pd
import numpy as np
import os
import glob
import re


# === Setup directory ===
results_dir = "/home/nandy.15/Research/Experiments_revision/Results/Pred_set_calibrate_mod_1"
csv_files = sorted(glob.glob(os.path.join(results_dir, "coverage_*.csv")))

records = []

# === Loop over all CSV files ===
for file in csv_files:
    # Extract lambda, N, and rho from filename
    match = re.search(r"coverage_lam([\d.]+)_N(\d+)_rho([\d.]+)\.csv", os.path.basename(file))
    if match:
        lam, N, rho = match.groups()
        lam = float(lam)
        N = int(N)
        rho = float(rho)

        df = pd.read_csv(file)

        records.append({
            "N": N,
            "lambda": lam,
            "rho": rho,
            "coverage": df["coverage"].values[0]  # assuming single row per file
        })

# === Create final DataFrame ===
final_df = pd.DataFrame(records)

# === Save to CSV ===
final_df.to_csv(os.path.join(results_dir, "combined_coverage_results.csv"), index=False)
print("Saved combined coverage results.")
```
